```python
##### 1.  
# 'RV'   :       ,         
credit_with_seg = credit_with_seg.drop(columns=['RV'])

# 'RV'  'RV'
# credit_with_seg['RV__'] = (credit_with_seg['RV'] == 'N').astype(int) : 'RV'  'N'  True,  False  , astype(int)  True 1, False 0 
# credit_with_seg = credit_with_seg.drop(columns='RV') : 'RV'      drop   
credit_with_seg['RV__'] = (credit_with_seg['RV'] == 'N').astype(int)
credit_with_seg = credit_with_seg.drop(columns='RV')

##### 2.  
# _R12M: '0'  0, '1'  1, '2'  2
# _R12M: str.replace('', '')  "1", "2"      int 
credit_with_seg['_R12M'] = (
    credit_with_seg['_R12M']
    .str.replace('', '', regex=False)  # '' 
    .astype(int)  #  int !
)

#_R12M  '_R12M_': '0'  0, '1'  1
# , : .map()  '0' 0, '1' 1   
credit_with_seg['_R12M_'] = credit_with_seg['_R12M'].map({
    '0': 0,
    '1': 1
    }).astype(int)
credit_with_seg.drop(columns=['_R12M'], inplace=True)

# '': 'Y'  1, 'N'  0
# :  'Y/N'   ,    1 0 
credit_with_seg[''] = credit_with_seg[''].map({
    'Y': 1,
    'N': 0
    }).astype(int)

# ''  '': '0'  0, '1'  1
credit_with_seg[''] = credit_with_seg[''].map({
    '0': 0,
    '1': 1
    }).astype(int)
credit_with_seg.drop(columns=[''], inplace=True)
```

```python
##### 3.  
#     1 
credit_with_seg = credit_with_seg.drop(columns=['_R3M'])

##### 4.   
# 'RV'  
credit_with_seg['RV'] = (credit_with_seg['RV'] > 0).astype(int)

# '_2'  
credit_with_seg['_2'] = (credit_with_seg['_R12M'] > 1).astype(int)

# '_R12M_3'  
credit_with_seg['_R12M_3'] = (credit_with_seg['_R12M'] > 2).astype(int)

# 'CA_1'  
credit_with_seg['CA_1'] = (credit_with_seg['CA'] == 1).astype(int)

# '_A'  
a_limit = credit_with_seg[credit_with_seg['Segment'] == 'A']['_B1M']
q1 = a_limit.quantile(0.25)
q3 = a_limit.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
a_no_outlier = a_limit[(a_limit >= lower_bound) & (a_limit <= upper_bound)]
a_B1M = a_no_outlier.min()
print(f"A    : {a_B1M:,}")  # 170,344

a_limit = credit_with_seg[credit_with_seg['Segment'] == 'A']['_B2M']
q1 = a_limit.quantile(0.25)
q3 = a_limit.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
a_no_outlier = a_limit[(a_limit >= lower_bound) & (a_limit <= upper_bound)]
a_B2M = a_no_outlier.min()
print(f"A    : {a_B2M:,}")  # 170,344

def classify_dual_limit(row):
    b1 = row['_B1M'] >= a_B1M
    b2 = row['_B2M'] >= a_B2M
    if b1 and b2:
        return 2     #   170,344 
    elif b1 or b2:
        return 1     #  170,344 
    else:
        return 0    #   170,344 

credit_with_seg['_A'] = credit_with_seg.apply(classify_dual_limit, axis=1)
```

```python
import pandas as pd
import numpy as np
import gc

# from google.colab import drive
# drive.mount('/content/drive')

# # LightGBM  GPU driver  
# !mkdir -p /etc/OpenCL/vendors && echo "libnvidia-opencl.so.1" > /etc/OpenCL/vendors/nvidia.icd
# !sudo apt install nvidia-driver-460 nvidia-cuda-toolkit clinfo
# !apt-get update --fix-missing
# !pip install -q  lightgbm==4.1.0 \
#   --config-settings=cmake.define.USE_GPU=ON \
#   --config-settings=cmake.define.OpenCL_INCLUDE_DIR="/usr/local/cuda/include/" \
#   --config-settings=cmake.define.OpenCL_LIBRARY="/usr/local/cuda/lib64/libOpenCL.so"

# # catboost 
# !pip install catboost

import sklearn
from sklearn.utils.class_weight import compute_class_weight
import imblearn
from imblearn.over_sampling import SMOTE
import xgboost as xgb
from xgboost import XGBClassifier
import catboost
from catboost import CatBoostClassifier
import lightgbm as lgb
from lightgbm import early_stopping, log_evaluation, LGBMClassifier

print("   !")
```

```python
test_df = pd.read_parquet("./data/test/test_df_cleaned.parquet")

#  500 
top500_df = pd.read_csv("./data/top500_features_XGB_balanced.csv")
top500_features = top500_df['feature'].tolist()
```

```python
#   
xgb_model_loaded = xgb.XGBClassifier()
xgb_model_loaded.load_model('./data/softvoting_xgb_xgb_3_3_25.json')

lgb_model_loaded = lgb.Booster(model_file='./data/softvoting_lgbm_model_3_3_25.txt')

cat_model_loaded = CatBoostClassifier()
cat_model_loaded.load_model('./data/softvoting_cat_xgb_6_12_24.cbm')
```

```python
# test  
X_test = test_df[top500_features]

# XGB predict_proba (iteration_range)
proba_xgb = xgb_model_loaded.predict_proba(X_test, iteration_range=(0, 5000))

# LGB Booster (predict proba)
proba_lgb = lgb_model_loaded.predict(X_test, num_iteration=3348)

# CAT predict_proba
proba_cat = cat_model_loaded.predict_proba(X_test)

# soft voting
ensemble_proba = (0.4 * proba_xgb) + (0.3 * proba_lgb) + (0.3 * proba_cat)
ensemble_preds = np.argmax(ensemble_proba, axis=1)
inverse_label_map = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E'}
ensemble_preds_label = pd.Series(ensemble_preds).map(inverse_label_map)
```

```python
#  test_df 
test_df["pred_label"] = ensemble_preds_label

# ID mode aggregation (   label)
submission = test_df.groupby("ID")["pred_label"] \
    .agg(lambda x: x.value_counts().idxmax()) \
    .reset_index()

#  
submission.columns = ["ID", "Segment"]

# 
submission.to_csv(
    "./data/submission_final.csv",
    index=False,
    encoding="utf-8-sig"
)

print("   !")
```
